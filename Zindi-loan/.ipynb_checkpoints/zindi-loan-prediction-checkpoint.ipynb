```python
# import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
import sklearn
```

```python
# load the data
train_demograph = pd.read_csv('data/train/traindemographics.csv') # demographics of customers
train_prev_loans = pd.read_csv('data/train/trainprevloans.csv') # previous loans of customers
train_perf = pd.read_csv('data/train/trainperf.csv')
```

Upon having a quick look at the datasets, we can see that there are a few columns that contain null values, and good_bad_flag column needs to be encoded according to the submission file. All datasets contain customer id, which can be used as a primary key to join the datasets.

```python
# first payment deadline met or not -> 0 (no), 1 (yes)
train_prev_loans['firstduedate']
train_prev_loans['firstrepaiddate']
train_prev_loans['deadlinemet'] = np.where(train_prev_loans['firstduedate'] > train_prev_loans['firstrepaiddate'], 1, 0)
train_prev_loans.head()
```

```python
# merge the datasets
# only merge deadlinemet column from previous loan dataset
train_temp = train_demograph.merge(train_prev_loans[['customerid', 'deadlinemet']], how='inner', on='customerid') # do we need previous loan data?
train = train_temp.merge(train_perf, how='inner', on='customerid') 
train.head()
```

```python
train.columns
```

Looking at the merged dataset, we need to decide which features to preserve or drop, replace null values, etc. First, we'll fill in the missing values.

```python
# explore the NaN values in the dataset
train.isnull().sum()
```

```python
# replace the nan values to "Unknown"
train['bank_branch_clients'] = train['bank_branch_clients'].fillna('Unknown')
train['employment_status_clients'] = train['employment_status_clients'].fillna('Unknown')
train['level_of_education_clients'] = train['level_of_education_clients'].fillna('Unknown')
```

```python
# replace referredby column with 0 or 1, where 0 is no, 1 is yes
train['referredby'] = train['referredby'].notnull().astype('int')
```

We'll then encode the good_bad_flag to be 0 or 1. 1 indicates "Good" while 0 indicates "Bad". We will use label encoding method by first converting the column from object to category data type, and then convert the string values to numerical using cat.codes built-in function.

```python
# convert good_bad_flag to numerical values via label encoding
if (train['good_bad_flag'].isnull().sum() == 0):
    train['good_bad_flag'] = train['good_bad_flag'].astype('category')
    print(train.dtypes)

train['good_bad_flag'] = train['good_bad_flag'].cat.codes # 1 = good, 0 = bad
train.head()
```

Let's explore this giant dataset. We'll visualize the dataset to explore the characteristics of the dataset.

```python
# visualization of train performance dataset
seaborn.countplot(train['good_bad_flag'])
```

```python
# does employment status matter?
seaborn.heatmap(train.corr())
```
